```python
import jieba
from urllib.request import urlretrieve
# 1. 載入大辭典(from結巴)
# 大辭典>精準度 ， 小辭典(詞彙量較少)>速度dict.txt.small
url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
urlretrieve(url, "dict.txt.big")
jieba.set_dictionary("dict.txt.big")
# 2. 載入自定義辭典 politics是自定義辭典>可以隨意取名
# 製作辭典：記事本>"詞彙" "詞性" > 成詞率jieba會幫你算平均值，可以不填
# 找專有名詞列表給記事本
url2 = "https://github.com/Elwing-Chou/tibaml1027/raw/main/politics.txt"
urlretrieve(url2, "politics.txt")
#已經有大辭典做底了，加上載入自定義辭典
jieba.load_userdict("politics.txt")
```

```python
# list(jieba.cut(article))
" ".join(jieba.cut(article)) #用空白鍵分詞>長得像英文
#上下文會影響分詞判斷
```

```python
#示範結巴分詞工具
import jieba.analyse
jieba.analyse.extract_tags(article)
#topK > 參數是表示想抓的詞彙數量
jieba.analyse.extract_tags(article, topK=None)
#withWeight > 表示該詞彙在文章內的分數占比
jieba.analyse.extract_tags(article, topK=None, withWeight=True)
#allowPOS > 定義分詞的詞性來做篩選,可以到官方文件看詞性
jieba.analyse.extract_tags(article, topK=None, allowPOS=["n", "nr", "ns"])
```
