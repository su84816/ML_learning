```python
import pandas as pd
from urllib.request import urlretrieve
url = "https://github.com/Elwing-Chou/tibaml1027/raw/main/poem_train.csv"
urlretrieve(url, "train.csv")
train_df = pd.read_csv("train.csv", encoding="utf-8")
url = "https://github.com/Elwing-Chou/tibaml1027/raw/main/poem_test.csv"
urlretrieve(url, "test.csv")
test_df = pd.read_csv("test.csv", encoding="utf-8")
```

```python
train_df
# test_df
```

```python
# 進到函式庫之前, 所有東西換成數字
# 李白, 杜甫, 白居易 -> 0, 1, 2
# Series.replace({"李白":0, "杜甫":1})
# Series.unique/value_counts >>> 印出欄位/記次數
u = train_df["作者"].unique()

# 這裡在建一個辭典,分別把作者標上索引值存回字典,後面用串列生成式簡寫
# name2idx = {}
# for i, n in enumerate(u):
#    name2idx[n] = i
# 你也可以試試看: [i ** 2 for i in range(5)]

#轉成數字演算法才能比對
name2idx = {n:i for i, n in enumerate(u)}
idx2name = {i:n for i, n in enumerate(u)}
y_train = train_df["作者"].replace(name2idx)
y_test = test_df["作者"].replace(name2idx)
y_test
```

```python
# Series.apply(func)
import jieba
# s = "百川日東流，客去亦不息。我生苦漂蕩，何時有終極。"
# " ".join(jieba.cut(s))
def poemcut(p):
    return " ".join(jieba.cut(p))
x_train = train_df["內容"].apply(poemcut)
x_test = test_df["內容"].apply(poemcut)
x_test
```

```python
#這裡用fit+transform做標準化
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
#以train來做標準>fit:找出最大&最小,transform:做標準化
x_train_count = vec.fit_transform(x_train)
#test不用再做fit,直接transform就好 (train已經fit過,依據train定義的fit去做tranform)
x_test_count = vec.transform(x_test)
```

```python
# vec.vocabulary_ #可以看出分詞的欄位位置
```

```python
# print(x_train_count) #這樣印可以看出全部
x_train_count #sparse matrix>>全部欄位有2731*52294筆,但很多都是0,實際上只有85677是有值的
```

```python
# 下面的在check標點和換行沒被算進去(。, \n, \r\n): KeyError才是對的
vec.vocabulary_["？"] #vec內找值 #如果有的話用filter清洗掉 或 用replace處理
#抓一個字會發現沒有被算進去,這個原因是token patten裡面的參數設定分詞的原則 #裡面參數是正規表達式
```

```python
from sklearn.naive_bayes import MultinomialNB #單純貝式分類
clf = MultinomialNB(alpha=0.5) #alpha是關鍵參數,代表保底數值0~1,通常是本值的1/100~1/1000，太大會不公平 , 預設是1(Laplace smoothing)
clf.fit(x_train_count, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_count)
accuracy_score(pre, y_test) #驗證準確率
```

```python
from sklearn.metrics import confusion_matrix
#對驗證做表格輸出,會拿到一個混淆矩陣
pd.DataFrame(confusion_matrix(y_test, pre),
             index=["{}(正確)".format(n) for n in u],
             columns=["{}(預測)".format(n) for n in u])
```

```python
#做一個分詞機率比對
#後面有e-11...代表10的-11次方 , 可以用round做四捨五入
testp = input("來首詩:")
testpcount = vec.transform([poemcut(testp)])
proba = clf.predict_proba(testpcount)[0]
for n, p in zip(u, proba):
    print(n, "的機率是:", p)
```
